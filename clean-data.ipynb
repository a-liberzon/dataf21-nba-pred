This notebook contains the code I used to clean up and merge the data.

```python
import numpy as np
import pandas as pd
```

```python
def to_year(x):
    return str(int(x[0:4])+1)
```

```python
# load the all_seasons.csv
df1 = pd.read_csv("data/all_seasons.csv")
df1 = df1.drop(columns=['order', 'college', 'country', 'draft_year', 'draft_round', 'draft_number'])
df1['season'] = (df1['season'].apply(to_year)).astype(int)
df1 = df1.drop(df1[ df1['season'] < 2000].index)
df1
```

```python
df2 = pd.read_csv("data/nba-salaries.csv")
df2 = df2.drop(columns=['rank','team']).rename(columns={'name' : 'player_name'})
df2
```

```python
df = pd.merge(df1, df2)
df.to_csv('data/out/season-data.csv')
```

```python
df3 = pd.read_csv("data/historical-nba-performance.csv")
df3['season'] = (df3['season'].apply(to_year)).astype(int)
df3 = df3.drop(df3[ df3['season'] < 2000].index)
df3.to_csv('data/out/team-records.csv')
```

```python
df4 = pd.read_csv('data/team-records-edit.csv')
```

```python
def to_tot_games(x):
    return int(x[0:2]) + int(x[3:5])
```

```python
df4['tot_games'] = (df4['record'].apply(to_tot_games))
df = pd.merge(df, df4)
df.to_csv('data/out/season-data-final.csv')  
```

```python
df = pd.read_csv('data/out/season-data-edit.csv')
df = pd.merge(df, pd.read_csv('data/final/nba-salary-cap.csv'))
df['GP%'] = df['GP']/df['TG']
df['Cap%'] = df['Salary']/df['Cap']
df = df.drop(df[ df['GP'] < 11].index)
df.to_csv('data/final/season-data-final-edit.csv')
```
